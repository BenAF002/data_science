## What is a Manifold?
A fairly lazy primer...

### Homeomorphism
First, we need to understand homeomorphisms. A homeomorphism is a ***Bijective*** and continuous function between topological spaces that has a continuous inverse function.

The mapping $\mathbf{\Phi}: \mathcal{V} \rightarrow \mathcal{W}$ is a homeomorphism if:
1. It is bijective
2. It is continuous
3. Its inverse $\mathbf{\Phi}^{-1}$ is also continuous

This requires a lot of unpacking...

***1. Bijective:***

Let $\mathcal{V}$ and $\mathcal{W}$ be vector sets, then $\mathbf{\Phi}: \ \mathcal{V} \rightarrow \mathcal{W}$ is a mapping between the sets. This mapping is:
- ***Injective*** if: $$\forall \ \mathbf{x}, \mathbf{y} \in \mathcal{V} \ : \ \mathbf{\Phi(x)} = \mathbf{\Phi(y)} \ \implies \mathbf{x} = \mathbf{y} \\ \text{the contrapositive is:} \ \forall \ \mathbf{x}, \mathbf{y} \in \mathcal{V} \ : \ \mathbf{\Phi(x)} \ne \mathbf{\Phi(y)} \ \implies \mathbf{x} \ne \mathbf{y}$$
    - Written out: a function is injective if it maps distinct elements in the domain of $\mathcal{V}$ to distinct elements in the codomain $\mathcal{W}$
    - A consequence of injectivity is that different inputs will *always* lead to different outputs
    - E.g. $f(x) = 2x$ is injective
    - E.g. $f(x) = x^2$ is **not** injective -- $f(2) = f(-2) = 4$
- ***Surjective*** if: $$\mathbf{\Phi}(\mathcal{V}) = \mathcal{W} \\ \ \\ \text{equivalently:} \ \forall \ \mathbf{y} \in \mathcal{W}, \ \exists \mathbf{x} \in \mathcal{V} \ : \ \mathbf{\Phi}(\mathbf{x}) = \mathbf{y}$$
    - Written out: a function is surjective if it maps points from the domain of $\mathcal{V}$ onto *all* points in the codomain $\mathcal{W}$
    - Geometrically, this means that the mapping  allows for no gaps or missing points
    - E.g. $f(x) = x^2$ is **not** surjective onf $\reals$ because it never reaches any negative numbers

A function (mapping) is **Bijective** if it is both injective and surjective.

### Onto manifolds...
For our purposes, an $k$D manifold may be considered as a $k$-dimensional surface *within* higher $n$-dimensional spaces. These surfaces are technically of the same dimensions as the space within which they lie, but they may be represented in lower dimensions through a homeomorphic mapping. Some good intuition for this may come from a line in $\reals^2$:

```python
x = np.arange(-1, 1, 0.01)
y = x**2
plt.plot(x, y);
```

This parabola exists in the 2-dimensional space of $\reals^2$ but it does not cover every point in $\reals^2$, it is itself a 1-dimensional line, and we may (somehow) map from $\reals^2$ onto it homeomorphically. Thus, it is a 1D-manifold.

This is the case for surfaces in higher dimensions. For instance, in $\reals^n$, any *non-intersecting* hyperplane is a 2D-manifold *embedded* in $\reals^n$. Why non-intersecting? Because at the point of intersection, it would no longer appear 2D - it would no longer be homeomorphic from $\reals^n$ to $\reals^2$.

More formal topological definition of manifolds: 

An $n$-dimensional topological manifold $M$ is locally homeomorphic to $\reals^n$. This means that for every point $\mathbf{p} \in M$ there is an *open* neighborhood $U$ of $\mathbf{p}$ and a homeomorphism $\phi: \ U \rightarrow V$ which maps the set $U$ onto an open set $V \subset \reals^n$
- The mapping $\phi: \ U \rightarrow V$ is called a ***chart*** or ***coordinate system***
- The set $U$ is the *domain* or *local coordinate neighborhood* of the chart
- The image of the point $\mathbf{p} \in U$, denoted $\phi(\mathbf{p}) \in \reals^n$ is called the ***coordinates*** or *local coordinates* of $\mathbf{p}$ in the chart
- A set of charts $\{\phi_\alpha | \alpha \in \mathbb{N}\}$ with domains $U_\alpha$ is called the ***atlas*** of $M$ if $\bigcup_{\alpha\in\mathbb{N}}U_\alpha = M$

Let's hold on this for a moment... \
We have a homeomorphic mapping $\phi$ from a set $U$ on the manifold $M$ into a set within $\reals^n$. So, basically, we can map the $n$-dimensional manifold in $n$-dimensional space through a coordinate system $\phi$. Kind of like how we charted the parabola in $\reals^2$ using the cartesian coordinate system...?

**NOTE:** \
 Within data science and deep learning conversations, we typically consider manifolds as lower dimensional surfaces embedded within higher-dimensional spaces. However, from this definition, they need not be embedded! They can actually be higher dimensional that the space within which we're working. They just have to homeomorphically map into the space...
