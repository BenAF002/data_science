## Geometric Dot Product - Dot Product is Defined by the Angle b/w Vectors

The angle between two vectors defines their dot-product. 

***Cosine Law:***
$$\|\vec{u} - \vec{v}\|^2 = \|\vec{u}\|^2 + \|\vec{v}\|^2 - 2\|\vec{u}\|\|\vec{v}\|\text{cos}(\theta)$$


Expanding the cosine law for vectors in $\mathbb{R}^3$:
$$\|\vec{u}\|^2 = u_1^2 + u_2^2 + u_3^2; \|\vec{v}\|^2 = v_1^2 + v_2^2 + v_3^2$$
$$\|\vec{u} - \vec{v}\|^2 = (u_1 - v_1)^2 + (u_2 - v_2)^2 + (u_3 - v_3)^2 $$

After expanding the binomials and cancelling like terms:
$$-2u_1u_2 - 2u_2v_2 - 2u_3v_3 = -2\|\vec{u}\|\|\vec{v}\|\text{cos}(\theta)$$

Then solving for $\theta$ which is the angle between the two vectors:
$$\theta = \text{cos}^{-1}\left(\frac{u_1v_1 + u_2v_2 + u_3v_3}{\|\vec{u}\|\|\vec{v}\|}\right) = \text{cos}^{-1}\left(\frac{\langle \vec{u}, \vec{v} \rangle}{\|\vec{u}\|\|\vec{v}\|}\right)$$

The numerator in the arcosine expression of the angle between the two vectors **is their dot product**.

## Geometric Cross Product

The cross product of two vectors in $\mathbb{R}^3$ yields the vector that is orthogonal to the two vectors. This orthogonal vector can inform us about the tilt of the *plane* spanned by the two original vectors in $\mathbb{R}^3$. The vector produced by the cross product has a couple other nifty features:
$$\vec{u} \times \vec{v} = \vec{n} $$
1. $\vec{n}$ is orthogonal to $\vec{u}$ and $\vec{v}$
2. $\vec{n}$ has length equal to the area of the parrallelogram enclosed by $\vec{u}$ and $\vec{v}$

## Vector Equation of Lines

In $\mathbb{R}^2$ we almost always think of lines in the cartesian coordinate system with the equation $$y - y_0 = m(x - x_0)$$
In higher dimensional spaces, like $\mathbb{R}^3$, a line can be defined by a point (through which the line passes) and a vector. The point, $P_0(x_0, y_0, z_0)$, is defined by a vector $\vec{r}_0$ radiating from the origin. The point lies at the tip of $\vec{r}_0$. Then vector $\vec{v}$ radiating from this point, spans the line. Then, *any other point on the line* may be defined by the addition of $\vec{r}_0$ and $t\vec{v}$ where $t$ is a scalar. That is, any point, $P(x, y, z)$, at the tip of some vector radiating from the origin, $\vec{r}$, is given by:
$$\vec{r} = \vec{r}_0 + t\vec{v}$$

## Equations of Planes

A plane in $\mathbb{R}^3$ can be defined with a vector that lies on the plane and a vector that is orthogonal to both the original vector *and the plane*. While we need two vectors that lie on a plane (and that are linearly independent of eachother) to span the plane, we actually can define a plane with only one vector that is on the plane and the vector that is orthogonal to the plane. This is becase any point on the plane can be described by a vector on the plane (i.e., its coordinates may be described by a vector on the plane), and the orientation (tilt) of the plane can be described by the orthogonal vector to the plane.

For some point on a plane, $P_0$, and the vector, $\vec{n}$, that is orthonormal to the plane, the equation for the plane is: $$\langle \vec{n}, \vec{P_0P} \rangle= 0$$

More explicitly, the equation of a plane through $P_0$ is the dot product of vectors $\vec{n}$ and $\vec{P_0P}$ where $\vec{n}$ and $\vec{P_0P}$ are chosen such that their dot product equals zero (i.e., such that they are orthogonal). The vector $\vec{P_0P} \equiv <x-p_{0,1}, y - p_{0,2}, z - p_{0,3}>$

This equation is the ***Vector Form*** of the plane. The ***Component Form*** of the plane equation is simply the dot product expanded out.

## Intro to Vector-Valued Functions

$$\vec{r}(t) = f(t)\hat{i} + g(t)\hat{j} + h(t)\hat{k}$$

A vector-valued function is a function with an **output** that is multidimensional and an **input** that is one-dimensional. The vector-valued function $\vec{r}(t)$ is defined in $\mathbb{R}^3$. It is a function of the one-dimensional input $t$, but its output is three-dimensional. The output may be decomposed as a linear combination of three functions of $t$ along the three basis vectors $\hat{i}$, $\hat{j}$, and $\hat{k}$; that is, in the directions of the three basis vectors that define the space over which the function $\vec{r}(t)$ is defined. These three functions ($f(t), g(t), and h(t)$) are called ***Scalar Functions***.


Let's look at an example: $$\vec{r}(t) = \text{cos}(t)\hat{i} + \text{sin}(t)\hat{j} + t\hat{k}$$

```python
fig = plt.figure()
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')

# input variable (independent variable)
t = np.linspace(0, 20, 100)

# defining all 3 axis
i = np.cos(t)
j = np.sin(t)
k = t
 
# plotting
ax.plot3D(i, j, k,'b')
ax.set_title('Helix')
plt.show()
```

***Tangent Vectors*** are vectors that are tangent to the vector-valued function. So, at each point, the tangent vector to that point describes the *direction of increase* in the vector-valued function. The tangent vector is the vectorized total derivative of the vector-valued function. The definition of the total derivative is analogous to that of ordinary functions:
$$\frac{d\vec{r}}{dt} = \lim_{\Delta t \to 0} \frac{\vec{r}(t+\Delta t) - \vec{r}(t)}{\Delta t}$$

This is the same as: $$\frac{d\vec{r}}{dt} = \frac{df}{dt}\hat{i} + \frac{dg}{dt}\hat{j} + \frac{dh}{dt}\hat{k}$$

For the example helix function above, the tangent vector to the helix is:
$$\frac{d\vec{r}}{dt} = -\sin(t)\hat{i} + \cos(t)\hat{j} + \hat{k}$$

## Arclength in 3D

How can we determine the length of a curve? This is the objective of ***Arclength***.

We can motivate arclength by considering how we might discretely approximate the length of a curve like the helix:

```python
fig = plt.figure()
 
ax = plt.axes(projection ='3d')

# input variable (independent variable)
t = np.linspace(0, 10, 100)

# defining all 3 axis
i = np.cos(t)
j = np.sin(t)
k = t
 
# approximations
aprx = np.linspace(0, 10, 10)
aprx_i = np.cos(aprx)
aprx_j = np.sin(aprx)
aprx_k = aprx
    
# plotting
ax.plot3D(i, j, k,'b')
ax.plot3D(aprx_i, aprx_j, aprx_k,'orange')
ax.set_title('Helix')
plt.show()
```

Here, we separated the helix defined for $0 \le t \le 10$ into ten equal length line-segment approximations. We can approximate the length of the helix over the interval of $t$ as:
$$Arclength \approx \sum_{i=1}^{10}(\text{Length of }i^{th}\text{ segment})$$

The approximation will converge to the true arclength as we add increasingly many, increasingly small line segments: $$Arclength \approx \lim_{n \to \infty} \sum_{i=1}^{n}(\text{Length of }i^{th}\text{ segment})$$

Now, the length of each line-segment may be computed directly with the distance formula. That is, as the sqrt of the sum of the squared changes in each direction. For segment $L_i$: $$\Delta L_i = \sqrt{(\Delta x_i)^2 + (\Delta y_i)^2 + (\Delta z_i)^2}$$
$$Arclength \approx \lim_{n \to \infty} \sum_{i=1}^{n}\sqrt{(\Delta x_i)^2 + (\Delta y_i)^2 + (\Delta z_i)^2}$$

As the line segment approximations become increasingly small, down to a limit of length zero:
$$Arclength \approx \lim_{n \to \infty} \sum_{i=1}^{n}\sqrt{\left(\frac{\Delta x_i}{\Delta t}\right)^2 + \left(\frac{\Delta y_i}{\Delta t}\right)^2 + \left(\frac{\Delta z_i}{\Delta t}\right)^2}\Delta t$$

Each of these terms in the square root are the derivatives in the $x$, $y$, and $z$ directions as $t$ approaches zero.

Furthermore, the limit of this sum as the number of elements approaches infinity is an integral over the domain of $t$. So, the full expression for the arclengthl, *without approximation* is:
$$Arclength = \int_{a}^{b} \sqrt{f'(t)^2 + g'(t)^2 + h'(t)^2}dt$$
For $\vec{r}(t) = f(t)\hat{i} + g(t)\hat{j} + h(t)\hat{k}$ with $a \le t \le b$

## Measuring Curvature

Think about tangent vectors for a moment. The direction of the tangent vector tells us the direction in which the curve is increasing. The magnitude of the tangent vector tells us how significant the increase is. What about the derivative? The derivative of the tangent vector describes how quickly the tangent vector is changing. Or, equivalently, the rate of change in the rate of change of the curve. Curvature describes how this rate of change changes - e.g. the steepness, peakedness, waviness, etc.

**Aside:** Saying "increase" isn't really appropriate. I should say the direction in which the curve is **changing**.

So, we can think of curvature as some relationship between how the tangent vector changes throughout the curve and the length of the curve (arclength).

For **unit** tangent vector $\vec{T}$, and arclength $s$, the curvature is defined as: 
$$\kappa = \left| \frac{d\vec{T}}{ds} \right|$$
In words, the curvature is the **magnitude** of the change in the tangent vector with respect to the change in the length of the curve.

This expression is a bit difficult to interpret (for me) because it's unclear what the derivative w.r.t. the arclength is. A more friendly expression is:
$$ \kappa = \frac{1}{\left| \vec{v} \right|} \left| \frac{d\vec{T}}{dt} \right|$$
Where $\vec{v}$ is the tangent vector. This expression is derived in the note on arclength below.

***Principle Unit Normal***
$$\vec{N} = \frac{1}{\kappa}\frac{d\vec{T}}{ds}$$
The principle unit normal, $\vec{N}$, is a vector that describes the direction in which the curve is curving at a given point. That is, $\vec{N}$ describes the direction in which the unit tangent vector, $\vec{T}$ will change as the arclength, $s$, changes. As per the name, the principle unit normal vector is a *normal* vector of the unit tangent vector. That is, it is *orthogonal to* the unit tangent vector.

***Circles of Curvature***
Defined as circles that:
1. Are tangent to the curve at a point
2. Have the same curvature as the curve (i.e. $\text{radius} = \frac{1}{\kappa}$)
3. Center on the concave (inner) side of the curve

The fundamental idea of the circle of curvature is that it is a circle with the same curvature as the underlying curve at a specific point. When the curvature of the curve is larger (i.e. the curve is more flat), then the circle of curvature will be larger when. When the curvature of the curve is smaller (i.e. the curve is more pronounced), then the circle of curvature will be smaller.

In this image, the blue line is the curve defined by the vector-valued function. The green circle is the **circle of curvature**. The yellow vector is the **tangent vector**. Finally, the orange vector is the **principle unit normal** to the tangent vector.

#### Note on Unit Tangent Vector

Let $\vec{v} = \frac{d\vec{r}}{dt}$. The unit tangent vector is: $$\vec{T} = \frac{\vec{v}}{\left|\vec{v}\right|}$$

#### Note on Arclength

Recall that arclength is given by: $$s = Arclength = \int \sqrt{f'(t)^2 + g'(t)^2 + h'(t)^2}dt$$

So, $$\frac{ds}{dt} = \sqrt{f'(t)^2 + g'(t)^2 + h'(t)^2}$$

For tangent vector $\vec{v}$, this is the magnitude of the tangent vector: $$\left| \vec{v} \right| = \frac{ds}{dt}$$

Therefore, if we expressed the curvature in term of the input variable $t$ instead of the arclength we would get:
$$\kappa = \left| \frac{d\vec{T}}{ds} \right| = \left| \frac{d\vec{T}}{dt} \frac{dt}{ds} \right| = \frac{1}{\left| \vec{v} \right|} \left| \frac{d\vec{T}}{dt} \right|$$

## Torsion

In $\mathbb{R}^3$ space, a curve's "motion" in the space can be fully described with three vectors, two of which we've already seen:
1. **Tangent Vector**: $$\vec{T} = \frac{d\vec{r}}{ds}$$
</br> - Describes the direction and magnitude of change in the curve (unit tangent vector only describes direction of change)
</br>

2. **Normal Vector**: $$\vec{N} = \frac{d\vec{T}}{ds}$$
</br> - Describes the direction and magnitude of change in the tangent vector (principle unit normal only describes direction of change)
</br> - Can interpret as the direction in which the curve is curving 
</br>

3. **Binormal Vector**: $$\vec{B} = \vec{T} \times \vec{N}$$
</br> - The cross product of the tangent vector and the normal vector
</br> - Describes how the curve ***twists*** in space

Let's take a look at the point on the helix where $t = 3.5$:

```python
fig = plt.figure()
 
ax = plt.axes(projection ='3d')

# input variable (independent variable)
t = np.linspace(0, 6, 100)

# defining all 3 axis
i = np.cos(t)
j = np.sin(t)
k = t
    
ax.plot3D(i, j, k,'b')    # full helix

t = 3.5
# point at t=3.5
ax.plot3D(np.cos(t), np.sin(t), t, color='orange', marker='.', markersize=10)    

# tangent vector at t=3.5
T = np.array([-np.sin(t), np.cos(t), 1])
ax.quiver(np.cos(t), np.sin(t), t,   # point coordinates
          T[0], T[1], T[2],          # vector coordinates
          color="orange")

# normal vector at t=3.5
mag_v = ((-np.sin(t))**2 + (np.cos(t))**2 + 1**2)**0.5    # magnitude of tangent vector
N = np.array([-np.cos(t)/mag_v, -np.sin(t)/mag_v, 0])
ax.quiver(np.cos(t), np.sin(t), t,    # point coordinates
          N[0], N[1], N[2],           # vector coordinates
          color="yellow")

# binormal vector at t=3.5
B = np.cross(T, N)
ax.quiver(np.cos(t), np.sin(t), t,    # point coordinates
          B[0], B[1], B[2],           # vector coordinates
          color="purple")

ax.set_title('Helix')
plt.show()
```

Here, the orange vector is the tangent vector, the yellow vector is the normal vector, and the purple vector is the binormal vector.

What does the change in the binormal vector tell us about the curve? The change in the binormal vector with respect to the curve's arclength is:
$$\frac{d\vec{B}}{ds} = \frac{d(\vec{T} \times \vec{N})}{ds}$$

The derivative of the cross product can be decomposed by the product rule:
$$\frac{d(\vec{T} \times \vec{N})}{ds} = \frac{d\vec{T}}{ds} \times \vec{N} + \vec{T} \times \frac{d\vec{N}}{ds}$$

This simplifies to: $$\frac{d(\vec{T} \times \vec{N})}{ds} =  \kappa \vec{N} \times \vec{N} + \vec{T} \times \frac{d\vec{N}}{ds} = \vec{T} \times \frac{d\vec{N}}{ds}$$
Because the cross product of a vector with itself is zero.

**Note on derivatives of unit vectors:** 

If $\left| \vec{r}(t) \right| = 1$ then, $$\langle \vec{r}'(t), \vec{r}(t) \rangle = 0$$
That is, the dot product of a vector-valued function and its derivative is 0 if the vector-valued function has a length of 1. So, the derivative is orthogonal to the vector-valued function. 

The derivative of the binormal vector of the unit tangent vector and the unit normal vector is the cross product of the unit tangent vector and the derivative of the unit normal vector w.r.t. the arclength. Thus, the binormal vector *is also a unit vector*. Therefore, the derivative of the binormal vector w.r.t. the arclength is orthogonal to the binormal vector. Explicitly:
1. $\frac{d\vec{B}}{ds} \perp \vec{T}$
2. $\frac{d\vec{B}}{ds} \perp \vec{B}$

Because $\frac{d\vec{B}}{ds}$ is orthogonal to $\vec{T}$ and $\vec{B}$ it **must be parrallel** to $\vec{N}$. Because it is parrallel to $\vec{N}$, and it passes through the same point as $\vec{N}$, we may express the derivative of the binormal vector w.r.t. the arclength as a scaled normal vector:
$$-\frac{d\vec{B}}{ds} = \tau \vec{N}$$
The negative sign is added by convention (it doesn't really matter...)

Taking the dot product of both sides of this expression with $\vec{N}$ removes the $\vec{N}$ term from the right-hand-side (because the dot product of a vector with itself is 1). So,
$$-\frac{d\vec{B}}{ds} \cdot \vec{N} = \tau$$

***Torsion*** is the scalar value $\tau$

Torsion describes the degree to which the curve **twists** in space. While $\kappa$ or *curvature* describes how much points curve on a plane defined by (spanned by) $\vec{T}$ and $\vec{N}$; the torsion, $\tau$ describes how much that plane curves or *twists* in space.

## Tangential and Normal Components of Acceleration

Velocity is given by the tangent vector. Explicitly, for unit tangent vector $\vec{T}$, the velocity vector is: $$\vec{v} = \left| \vec{v} \right| \vec{T}$$
Acceleration is the derivative of velocity w.r.t. $t$, which may be expanded as:
$$ \vec{a} = \frac{d\vec{v}}{dt} = \frac{d\left| \vec{v} \right|}{dt}\vec{T} + \left| \vec{v} \right| \frac{d\vec{T}}{dt} = \frac{d\left| \vec{v} \right|}{dt}\vec{T} + \kappa \left| \vec{v} \right|^2 \vec{N}$$

These coefficients to $\vec{T}$ and $\vec{N}$ are denoted:
$$\vec{a} = a_{\vec{T}}\vec{T} + a_{\vec{N}}\vec{N}$$

$a_{\vec{T}}$ is the ***Tangential Component*** of acceleration and $a_{\vec{N}}$ is the ***Normal Component*** of acceleration.

So, the acceleration $\vec{a}$ can be decomposed into a tangential component, $a_{\vec{T}}$, in the unit tangent direction, $\vec{T}$, and a normal component, $a_{\vec{N}}$, in the unit normal direction, $\vec{N}$.

One reason why this decomposition of acceleration is useful is that the curvature $\kappa$ is often difficult to compute. By computing the unit tangent vector, $\vec{T}$, and the acceleration, $\vec{a}$ we can find the curvature and the unit normal vector with:
</br>
$$\left| \vec{a} \right| = \sqrt{\left|a_{\vec{T}}\right|^2 + \left|a_{\vec{N}}\right|^2}$$
</br>
$$\left|a_{\vec{N}}\right|^2 = \left| \vec{a} \right|^2 - \left|a_{\vec{T}}\right|^2$$
</br>
$$\kappa = \frac{\left| \vec{a} \right|^2 - \left|a_{\vec{T}}\right|^2}{\left|\vec{v}\right|^2}$$
Note that neither $\vec{T}$ nor $\vec{N}$ are included in the expression for $\left|\vec{a}\right|$. This is because they are both unit vectors (they both have lengths of one).

## Limits of Multi-Variable Functions

Consider the function:
$$f(x,y) = \frac{xy}{x^2 + y^2}$$

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)

x, y = np.meshgrid(x, y)
z = (x*y) / (x**2 + y**2)

ax.plot_surface(x, y, z, cmap='winter')
plt.show()
```

This function is undefined at (0,0)! So, what is the limit at (0,0)? As we can see, the graph gets wonky as it approaches this limit, forming a sort of valley. With a little extra work, we can visualize exactly what is happening in the figure by plotting each region over which $x$ and $y$ are defined; that is, everywhere except for where $x$ and $y$ are **both** equal to 0.

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x0 = np.linspace(-2, -0.001, 200)
y0 = np.linspace(-2, -0.001, 200)

x1 = np.linspace(0.001, 2, 200)
y1 = np.linspace(0.001, 2, 200)

def f_z(x, y):
    z = (x*y) / (x**2 + y**2)
    return z

x0, y0 = np.meshgrid(x0, y0)
x1, y1 = np.meshgrid(x1, y1)

ax.plot_surface(x0, y0, f_z(x0, y0), cmap='winter')    # x < 0, y < 0
ax.plot_surface(x1, y1, f_z(x1, y1), cmap='winter')    # x > 0, y > 0
ax.plot_surface(x1, y0, f_z(x1, y0), cmap='winter')    # x > 0, y < 0
ax.plot_surface(x0, y1, f_z(x0, y1), cmap='winter')    # x < 0, y > 0

# ax.plot_surface(0, y1, f_z(0, y1), cmap='winter')      # x = 0, y > 0
# ax.plot_surface(0, y0, f_z(0, y1), cmap='winter')      # x = 0, y < 0
# ax.plot_surface(x1, 0, f_z(x1, 0), cmap='winter')      # x > 0, y = 0
# ax.plot_surface(x0, 0, f_z(x0, 0), cmap='winter')      # x < 0, y = 0

x, y = np.linspace(-2, 2, 400), np.linspace(-2, 2, 400)
ax.plot(x, y, 0.5, color='red')
ax.plot(-x, y, -0.5, color='red')
plt.show()
```

If we restrict the function to the line $y = x$, then the function simplifies to:
$$f(x) = \frac{x^2}{2x^2}$$
The limit of which is:
$$\lim_{x \to 0} \frac{x^2}{2x^2} = \frac{1}{2}$$

If we restrict the function to the line $y= -x$ instead, then the function simplifies to:
$$f(x) = \frac{-x^2}{2x^2}$$
The limit of which is:
$$\lim_{x \to 0} \frac{-x^2}{2x^2} = \frac{-1}{2}$$

These are the (mostly hidden) red lines in the figure. **The limit of the function $z$ is different as it is approached from different directions**.

This is called the ***Two Path Test*** for a limit:
- If $f(x,y)$ has two different limits along two different paths $(x,y) \to (x_0,y_0)$, then the limit $\lim_{(x,y) \to (x_0,y_0)} f(x,y)$ **does not exist**

## Computing Multivariable Limits Algebraically

How can we know if a limit *does exist* when we cannot check the value of the limit from *all* conceivable angles?

**Limit Laws:**

$$\lim_{(x,y) \to (a,b)} f(x,y) = L$$
$$\lim_{(x,y) \to (a,b)} g(x,y) = M$$
1. Limits for functions of the same inputs are additive:
</br>
$$\lim_{(x,y) \to (a,b)}\left(f(x,y) + g(x,y)\right) = L + M $$
2. Limits for functions of the same inputs are multiplicative:
</br>
$$\lim_{(x,y) \to (a,b)}\left(f(x,y) * g(x,y)\right) = L * M $$
$$\lim_{(x,y) \to (a,b)}\frac{f(x,y)}{g(x,y)} = \frac{L}{M} $$
3. Limits are scalable:
$$\lim_{(x,y) \to (a,b)}kf(x,y)= kL $$

A handy bit of trickery: If there exists a function $g(x)$ that is *continuous* at $L$, then:
$$\lim_{(x,y) \to (a,b)} g(f(x,y)) = g(L)$$

So, if we can evaluate a continuous function at the limit then we can compute the limit as the inverse of the function at that point.

## Partial Derivatives

By fixing all independent variables in a multivariable function to be constant except for one, we can examine the one-dimensional *slice* of the function at those fixed values. E.g., $$z = f(x,y) = -(x^2 + y^2)$$

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xm, ym = np.meshgrid(x, y)
def f_z(x, y):
    z = -(x**2 + y**2) + 1
    return z
    
ax.plot_surface(xm, ym, f_z(xm, ym), cmap='winter')

z = np.linspace(-7,1,100)
xm2, zm = np.meshgrid(x, z)

ax.plot_surface(xm2, -1, zm, alpha=0.4)

y_fixed = -np.ones(100)
ax.plot(x, y_fixed, f_z(x, y_fixed), color='red', zorder=5)

plt.show()
```

The red line is the function when we fix $y = -1$ for all values of $x$: $$f(x, y_0) = -(x^2 + (-1)^2)$$
The orange plane is the plane that function is restricted to with a fixed $y$. 

Now, the derivative of the constrained function w.r.t. $x$ is the ***Partial Derivative*** of the *unconstrained* function w.r.t. $x$.
$$\frac{\partial f}{\partial x} = \frac{d}{dx}f(x,y_0) = -2x$$

The partial derivative w.r.t. $x$ describes how the function changes as $x$ changes and **only** $x$ changes. So, for this two variable function, we can examine the independent effects of changes in $x$ or $y$. In other words, the partial derivatives of $x$ and $y$ describe the rates of change in the $x$ direction and $y$ direction of the function. 

This begs the question, what about the rates of change in **other directions**? Say, the $y=x$ direction? hmmm....

## Continuity vs. Partial Derivatives vs. Differentiability

A function is continuous (be it single or multivariable) if all limits over the domain of its inputs exist and those limits are equal to the function. In the single-variable case, discontinuous functions are *not differentiable*. They are not smooth, we cannot compute their derivative at the point(s) of discontinuity. In the multivariable case, discontinuous functions may have valid partial derivatives. Take $f(x,y) = 1, \forall xy = 0$ and $f(xy) = 0, \forall xy \ne 0$

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
y = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
xm, ym = np.meshgrid(x,y)

def f_z(x,y):
    z = x*y
    return (z==0)*1

ax.plot_surface(xm, ym, f_z(xm,ym))
ax.plot(np.zeros(9), y, f_z(np.zeros(9),y), color='r', zorder=4)         # x=0
ax.plot(x, np.zeros(9), f_z(x,np.zeros(9)), color='orange', zorder=5)    # y=0
plt.show()
```

This is a step function equal to 1 for $xy=0$ and 0 for $xy \ne 0$. The limit of the function along either axis is 1, but the limit from any other direction, like $y=x$ is 0. So the function is discontinuous. However, its partial derivatives still exist. Along either axis, so in the $x$ direction and the $y$ direction but only at the axis (i.e. not parrallel to it), the partial derivatives w.r.t. $x$ and $y$ are 0 and 0. 

So, this function is *discontinuous* at all points $(x,0), (0,y), (0,0)$ but has partial derivatives that **exist at the points of discontinuity**. What does this mean for the *differentiability* of the function? Is it differentiable??

## Differentiability of Multivariable Functions

Partial derivatives only describe the effect of a change in one variable while keeping the other variable(s) constant. Geometrically, they only capture the change in one direction, not multiple directions simultaneously. To consider the general differentiability of a multivariable function we need to consider the derivative in multiple simultaneous directions. That is, as all input variables change. So, formally, a function $f(x,y)$ is differentiable at a point $(x_0, y_0)$ if:
$$f(x_0 + \Delta x, y_0 + \Delta y) - f(x_0, y_0) = \frac{\partial f}{\partial x} \Delta x + \frac{\partial f}{\partial y} \Delta y + E_1(\Delta x) + E_2(\Delta y) $$
Where $E_1(\Delta x)$ and $E_2(\Delta y)$ are error terms such that: $$\lim_{\Delta x \to 0} \frac{E_1(\Delta x)}{\Delta x} = 0; \lim_{\Delta y \to 0} \frac{E_2(\Delta y)}{\Delta y} = 0$$

Let's return to the example step function plotted above. At the point $(0,0)$ we saw that both partial derivatives are $0$. How about the other terms:

Because $f(x,y)=0; \forall xy \ne 0$:
$$f(0 + \Delta x, 0 + \Delta y) = 0$$
Because $f(x,y)=1; \forall xy = 0$:
$$f(0,0) = 1$$
So, $$0 - 1 = 0 + 0 + E_1(\Delta x) + E_2(\Delta y)$$
These error terms must be large to sum to 1; therefore, the limits to $0$ of their $\Delta x$ and $\Delta y$ quotients cannot be equal to $0$. So, the function **is not differentiable** at $(0,0)$.

***Theorem:***

If $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$ are **continuous** on an open region $\mathbb{R}$, then $f$ is **differentiable** on $\mathbb{R}$.

This is a useful shorthand for checking differentiability.

## Multivariable Chain Rule

Single variable chain rule:
$$\frac{d}{dx} f(g(x)) = f'(g(x))g'(x)$$
$$\frac{df}{dx} = \frac{df}{dg}\frac{dg}{dx}$$


Multivariable chain rule:
$$w = f(x(t), y(t))$$

$$\frac{dw}{dt} = \frac{\partial w}{\partial x}\frac{dx}{dt} + \frac{\partial w}{\partial y}\frac{dy}{dt}$$

The total change in $w$ w.r.t. the underlying independent variable $t$ is the sum of the partial changes in $w$ w.r.t. the functions of $t$: $x(t)$ and $y(t)$. Which are themselves changed by a change in $t$ (ala. their derivatives w.r.t. $t$).

In other words, the total change in $w$ w.r.t. the underlying independent variable $t$ is *the result of* the distinct changes in $x(t)$ and $y(t)$ due to the change in $t$.

Note that we can think about this function $w$ as a vector-valued function where $x(t)$ and $y(t)$ are input vectors. The change in the output vector $w$ is the sum of the changes in the input vectors. **bam**

## Directional Derivatives

Let's establish a fixed direction with a unit vector:

$$\vec{u} = (u_1, u_2); \text{  where} \left|\vec{u}\right| = 1$$

Then we can paramaterize the input variables $x$ and $y$ along the direction in which $\vec{u}$ is pointing as:

$$x(s) = x_0 + su_1,\text{    } y(s) = y_0 + su_2$$

We can define the **directional derivative** in the direction of $\vec{u}$ as:

$$D_{\vec{u}} f(x_0, y_0) = \lim_{s \to 0} \frac{f(x_0 + su_1, y_0 + su_2) - f(x_0, y_0)}{s}$$

Both $x$ and $y$ are dependent on the scalar $s$ along the direction of $\vec{u}$. So we can use the 
chain rule: 

$$\frac{d}{ds}f(x(s),y(s)) = \frac{\partial f}{\partial x}\frac{dx}{ds} + \frac{\partial f}{\partial y}\frac{dy}{ds}$$

Which is equivalent to:
$$D_{\vec{u}}f(x_0,y_0) = \frac{\partial f}{\partial x}u_1 + \frac{\partial f}{\partial y}u_2$$

These partial derivatives in each separate direction ($x$ and $y$) can be compiled into the ***Gradient*** of $f$:

$$\nabla f = \langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \rangle$$

**Aside:** The $\langle, \rangle$ notation *does not* denote a dot product here, it is simply the brackets denoting a vector that Trefor uses in his videos.

So, this directional derivative is equal to the **dot product** of the gradient and the directional vector $\vec{u}$:
$$D_{\vec{u}}f(x_0, y_0) = \nabla f \cdot \vec{u}$$

If we can compute the partial derivatives in the inputs $x$ and $y$ (i.e. if we can find the directional derivatives in the $x$ and $y$ directions), then we can find the directional derivative in **any** direction by computing the dot product of the gradient composed of the $x$ and $y$ partials and the unit vector that points in that direction.

## Geometric Meaning of Gradient Vector

The gradient points in the direction of **greatest change** in the function. 

Let's revisit the directional derivative and rewrite it to include the angle, $\theta$, between the gradient and unit direction vector:

$$D_{\vec{u}}f(x_0, y_0) = \nabla f \cdot \vec{u} = \left|\nabla f \right|\left|\vec{u}\right|\cos\theta$$

This equivalency results from the relation of a dot product of two vectors to the angle between the two vectors.

From this equivalency, we can see that the **magnitude** of $\nabla f \cdot \vec{u}$ is **smallest** at $\theta = \pi / 2$ because $\cos\frac{\pi}{2} = 0$. So, the smallest directional derivative is in the direction that is perpendicular to the gradient. This occurs along the tangent lines to the contour lines of the surface. This should make sense because the contour lines are **level**, i.e. they are constant values of the output variable.

We can see that the gradient is in fact the direction of greates change beause the directional derivative is maximized when $\cos\theta = 1$, which occurs at $\theta = 0$.

## Tangent Planes

Consider again the two variable function $z = f(x,y)$. At some point $(x_0, y_0)$, the tangent vectors in the $x$ and $y$ directions are given by the partial derivatives of the function w.r.t. $x$ and $y$. These tangent vectors **span** the ***Tangent Plane*** of $f(x,y)$ at the point $(x_0, y_0)$.

```python
fig = plt.figure()
ax = plt.axes(projection="3d")

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
xm, ym = np.meshgrid(x, y)

def f(x, y):
    z = 2 - (x**2 + y**2)
    return z
    
ax.plot_surface(xm, ym, f(xm, ym), cmap='winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

x0, y0 = 0.5, -0.5
ax.plot(x0, y0, f(x0,y0), marker='.', zorder=5)

def f_x(x): return -2*x
def f_y(y): return -2*y

x_fixed = np.array([x0 for _ in range(100)])
y_fixed = np.array([y0 for _ in range(100)])

ax.plot(x_fixed, y, f(x_fixed, y),        # f(x0, y) equation
        color='plum', zorder=2.5)
ax.plot(x_fixed[:80], y[:80],             # domain of y and fixed x
        f_y(y0)*y[:80] + 2,               # equation of tangent line with slope f_y(y0)
        color='magenta', zorder=4)

ax.plot(x, y_fixed, f(x, y_fixed),        # f(x, y0) equation
        color='lightsalmon', zorder=2.5)
ax.plot(x[20:], y_fixed[20:],             # domain of x and fixed y
        f_x(x0)*x[20:] + 2,               # equation of tangent line with slope f_x(x0)
        color='orangered', zorder=4)

plt.show()
```

The plane spanned by these two tangent vectors is the tangent plane. The formula for the tangent plane to $(x_0, y_0)$ in $\mathbb{R}^3$ is: 

$$f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0) + z_0 = z$$

## Optimization & The Second Derivative Test

**First Derivative Test:**

If $f(x,y)$ has a local max/min at $(a,b)$, then $f_x(a,b) = f_y(a,b) = 0$.
- That is, the partial first derivatives must all equal zero at the point of a max/min

**Second Derivative Test:**
If the $1^{st}$ and $2^{nd}$ partial derivatives all exist on an open disk centered at $(a,b)$, then:
- When $f_x(a,b)=f_y(a,b)=0$:
    - $f$ has a local MAX if $f_{xx} < 0$ **and** $f_{xx}f_{yy} - f_{xy}^2 > 0$
    - $f$ has a local MIN if $f_{xx} > 0$ **and** $f_{xx}f_{yy} - f_{xy}^2 > 0$
    - $f$ has a local SADDLE if $f_{xx}f_{yy} - f_{xy}^2 < 0$
    - We cannot conclude anything about the local max/min if $f_{xx}f_{yy} - f_{xy}^2 = 0$
    
Note that we only need to look at the second partial derivative w.r.t. one input variable ($x$ here) to determine local maxima/minima. This is because the second condition, $f_{xx}f_{yy} - f_{xy}^2 > 0$, qualifies that the point is not a saddle so it will be a maxima/minima regardless of which direction we approach it from. 

***Mixed Partial Derivatives:***

$f_{xy}$ is a mixed partial derivative. It is given by: $$ f_{xy} = \frac{\partial}{\partial y}\frac{\partial f}{\partial x}$$

Including the mixed partial evaluates the concavity (i.e. whether the function is increasing or decreasing) from directions **other than** $x$ and $y$. When $f_{xx}f_{yy} - f_{xy}^2 > 0$, the second partial derivatives in both the $x$ and $y$ directions must **both** be positive or **both** be negative. In addition, the mixed partial derivatibve $f_{xy}$ must not be too **large**.

**Method:**
1. Find critical points using the first derivative test
2. Classify critical points as maxima/minima/saddles or inconclusive via the second derivative test

## Optimization with Boundaries

Goal is to optimize a function on a **closed, bounded region**. When we have a constraint over which we must optimize a function, say $f(x,y)$, we must consider the maxima and minima along the boundary of the constraint as well as the general maxima and minima. 

**Step 1: Find Critical Points**
$$f_x = 4x = 0 \Rightarrow x = 0$$
$$f_y = 2y - 1 = 0 \Rightarrow y = \frac{1}{2}$$
So, $(0, 1/2)$ is the **only** critical point. $f(0,1/2) = (0, -1/4)$ is the **candidate** local minimum value.

**Step 2: Classify Critical Points**
$$f_{xx} = 4 > 0$$
$$f_{xx}f_{yy} - f_{xy}^2 = 4(2) - 0 = 8 > 0$$

So, $(0, -1/4)$ is a local minimum.

**Step 3: Analyze the Boundary**
Here the boundary is given by:
$$x^2 + y^2 = 1$$

This boundary can be represented in the original function by **parameterizing** the function. From the unit circle: $$\cos^2\theta + \sin^2\theta = 1$$
So, we can set $x = \cos\theta$ and $y = \sin\theta$

$$f(\cos\theta, \sin\theta) = f(\theta) = 2\cos^2\theta + \sin^2\theta - \sin\theta$$

Now, we have represented the multivariable function $f(x,y)$ as a single variable function of $\theta$ along the boundary. So, we can find critcal points along this single variable function $f(\theta)$ by taking its derivative:

$$f'(\theta) = -2\cos(\theta)\sin(\theta) - \cos(\theta)$$

Which is equal to $0$ at $\theta = (\frac{\pi}{2}, \frac{3\pi}{2}, \frac{7\pi}{6}, \frac{11\pi}{6})$

Because the function $f(\theta)$ is constrained, we can actually just evaluate the function at each of these critical points, the largest values are maxima and the smallest values are minima.

Alternatively, we could set $x$ in terms of $y$ and solve using **direct substitution** rather than parameterization.

**Extreme Value Theorem:**

If $f: \mathbb{R}^n \to \mathbb{R}$ is continuous on a closed, bounded region $R$, then $f$ attains a maximum and minimum on $R$.

**Note:** $f: \mathbb{R}^n \to \mathbb{R}$ means that $f$ is a multivariable function mapping $n$ input variables to a single output variable.

In these bounded scenarios there are two possibilities for extrema:
1. The max/min are in the interior of $R$
2. The max/min are on the boundary of $R$
This is why we must check the boundary for extrema.

## Lagrange Multipliers

Optimize:

$$f(x,y)=xy + 1$$

Subject to the constraint:

$$g(x,y)=x^2+y^2 - 1 = 0$$

```python
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
xm, ym = np.meshgrid(x, y)

def f(x, y): return x*y + 1             # f(x,y)
def cst(x,y): return x**2 + y**2 - 1    # constraint

ax.plot_surface(xm, ym, f(xm, ym), cmap='winter', alpha=0.45)
ax.contour(xm, ym, f(xm, ym), 12, cmap='autumn_r')

# plot constraint on surface
xc = np.linspace(-1, 1, 100)    # domain of constraint is ([-1, 1], [-1, 1])
yc = np.sqrt(1 - xc**2)
ax.plot(xc, yc, f(xc,yc), color='red', zorder=5)
ax.plot(xc, -yc, f(xc,-yc), color='red', zorder=5)

# maxima on constraint boundary
p = 1/np.sqrt(2)
ax.plot(p, p, f(p, p), marker='.', color='g', zorder=10)
ax.plot(-p, -p, f(p, p), marker='.', color='g', zorder=10)

plt.show()
```

The maxima and minima along the constraint boundary are the points at which the contour lines are tanget to the boundary **bam**. That is, the points which share a tangent line with a contour line. Furthermore, the normal vectors to these shared tangent lines are linearly dependent; i.e., they overlap and are scalar multiples of  eachother. The normal to the tangent line of the contour is the gradient of the function $\nabla f$, while the normal to the tangent line of the boundary may be denoted $\nabla g$. Explicitly:
$$\nabla g = \lambda\nabla f$$

Where $\lambda$ is some constant scalar.

Think about the intuition for this relationship. At the point where $\nabla g = \lambda \nabla f$, the gradient of $f$ is **orthogonal to** the tangent line of the constraint function. Therefore, **no component** of $\nabla f$ points alongthe direction of change in the constraint function. Thus, the underlying function $f$ **cannot be increased or decreased** by moving from that point! **BAM**

***Lagrange Multiplier Method:***

Simultaneously solve: $$\nabla f = \lambda \nabla g$$ $$g(\cdot) = 0$$

So, find the gradient of the function, $\nabla f$ and solve the constraint $g(\cdot)$. Solving the constraint basically amounts to determining the domain of inputs for which the constraint is met (e.g., observing that $x \in [-1, 1]$ and $y \in [-1, 1]$ in $g(x,y) = 0$), and then determining the range of the original function $f(\cdot)$ over that domain.

Then, find the gradient of the constraint boundary (i.e., the range of $f(\cdot)$ over the domain of the constrained inputs), $\nabla g$, that is a scalar multiple of $\nabla f$.

$$f(x,y) = xy + 1 \Rightarrow \nabla f = \langle x,y\rangle$$
$$g(x,y) = x^2 + y^2 -1 \Rightarrow \nabla g = \langle 2x, 2y \rangle$$

$$ \nabla f = \lambda \nabla g $$

$$\Rightarrow
  \begin{bmatrix}
    y\\
    x
  \end{bmatrix}
    = \lambda
  \begin{bmatrix}
    2x\\
    2y
  \end{bmatrix}
$$

System of equations is then comprised of the gradient equivalency and the constraint function:
$$y = \lambda 2x$$
$$x = \lambda 2y$$
$$x^2 + y^2 - 1 = 0$$

So, we have **three** equations for 3 unknowns.

$$y = \lambda 2(\lambda 2y) = 4\lambda^2y \Rightarrow \lambda = (0, -1/2, 1/2)$$
$$\Rightarrow y = \pm x \Rightarrow x = \pm \frac{1}{\sqrt{2}}$$

So, we are left with the four points: 
$$\left( \frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right), 
\left( -\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}\right),\\
\left( -\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}\right),
\left( \frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}\right)$$

The top two points are the maxima of the boundary while the lower two are the minima.

## Lagrange Multipliers with 2 Constraints

Optimize: $$f(x,y,z)$$

Subject to the constraints: $$g_1(x,y,z)=0$$ $$g_2(x,y,z)=0$$

The gradients of these two constraints, $\nabla g_1$ & $\nabla g_2$, span a plane in $\mathbb{R}^3$. Where in the single constraint case, optimal $\nabla f$ was shown to be parrallel to the gradient vector of the single constraint; in the two constraint case, optimal $\nabla f$ is parrallel to the plane spanned by the two gradient vectors. That is, $\nabla f$ is a linear combination of the two gradient vectors: 

$$\nabla f = \lambda \nabla g_1 + \mu \nabla g_2$$

So the complete system of equations is:
$$\nabla f = \lambda \nabla g_1 + \mu \nabla g_2$$
$$g_1 = 0 \\ g_2=0$$

## Hessian Matrix

The matrix $H$ comprised of all of the *second* partial derivatives of a multivariable function $f$. Ex:
$$f(x,y) = e^{x/2}sin(y)$$

$$
H =
\begin{bmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{bmatrix} 
= 
\begin{bmatrix}
\frac{1}{4}e^{x/2}\sin(y) & \frac{1}{2}e^{x/2}\cos(y)\\
\frac{1}{2}e^{x/2}\cos(y) & -e^{x/2}\sin(y)
\end{bmatrix}
$$


We can use the elements of $H$ to evaluate the second derivative test. Note that for $f(x,y)$, $f_{xx}f_{yy} - f_{xy}^2$ is the **determinant** of $H$. 

The Hessian helps generalize the second partial derivative test to higher dimensional functions. But to talk about these higher dimensional functions we need some new notation. The conditions of the second derivative test are called ***Principles*** and are denoted $\Delta_i$. For functions of two variables we have only:

$$\Delta_1 : \{f_{xx} > 0, f_{xx} < 0\}$$
$$\Delta_2 : f_{xx}f_{yy} - f_{xy}^2 = \text{det}(H_2) \Rightarrow \{\text{det}(H_2) > 0, \text{det}(H_2) < 0, \text{det}(H_2) = 0\}$$

Where $H_2$ is the **upper-left-hand** [2x2] submatrix of $H$. For $f(x,y)$, $H_2 = H$.

The higher-order principles are the determinants of the [$i$x$i$] submatrices of $H$. Ex:

$$
f(x,y,z) \Rightarrow H = 
\begin{bmatrix}
f_{xx} & f_{xy} & f_{xz} \\
f_{yx} & f_{yy} & f_{yz} \\
f_{zx} & f_{zy} & f_{zz}
\end{bmatrix}
$$

$$
\Delta_1 = f_{xx} \\
\Delta_2 = \text{det}(H_2) = f_{xx}f_{yy} - f_{xy}^2 \\
\Delta_3 = \text{det}(H)
$$

But what do the conditions mean for higher dimensional functions? Well, in some sense the pattern from the two-dimensional case may be extended. For $f(x,y, z)$, i.e. $f: \mathbb{R}^3 \rightarrow \mathbb{R}$, we have the conditions:
1. Local MIN if: $\Delta_1>0, \Delta_2>0, \Delta_3>0$
2. Local MAX if: $\Delta_1<0, \Delta_2>0, \Delta_3<0$

If **all** principles are positive, then we have a local minimum. If the first principle, $\Delta_1$, is negative and the following principles **alternate** positive to negative, then we have a local maximum.

For $f(x,y)$ we saw that $\Delta_2$ describes the increase/decrease in the function from directions other than $x$ and $y$ by including the mixed partial derivative $f_{xy}$. The higher-order principles, like $\Delta_3$ similarly describe the change of the function in higher dimensions by including more mixed partial derivatives. 

As with $f(x,y)$, for $f(x,y,z)$ we also have the third condition:

3. Saddle point if $\Delta_2 < 0$

This condition holds *regardless of* the values of the higher-order principles.

Finally, we have:

4. Inconclusive if $\Delta_2 = 0$
